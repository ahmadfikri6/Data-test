```python
import pyttsx3
import json
engine = pyttsx3.init() # object creation
```

```python
# """ Kecepatan Bicara """
rate = engine.getProperty('rate')   # mengambil kecepatan suara
print (rate)                        # menampilkan kecepatan suara
engine.setProperty('rate', 200)     # mengubah kecepatan suara
```

```python
# """ Volume Suara """
volume = engine.getProperty('volume')   # mengambil level volume (min=0 and max=1)
print (volume)                          # menampilkan volume
engine.setProperty('volume',5.0)        # mengubah mengubah volume
```

```python
# """ Mengubah Suara """
voices = engine.getProperty('voices')       # mengambil detail jenis suara
#engine.setProperty('voice', voices[0].id)  # jika ingin mengubah menjadi pria
engine.setProperty('voice', voices[1].id)   # jika ingin mengubah menjadi wanita
```

```python
engine.say("Apa kau liat liat")
engine.runAndWait()
engine.stop()
```
